```python
import csv
import pandas as pd
import matplotlib.pyplot as plt
import json

#Student File
file01 = open("Student.csv", "x+")

sd = csv.writer(file01)
studat = []
for i in range(0,9):
    n1 = input('Student ID :')
    n2 = input('Student Name :')
    n3 = int(input('Student Roll :'))
    n4 = input('Batch ID')
    studat.append(
        {
            'Student_ID': n1,
            'Student_Name': n2,
            'Roll_Number':  n3,
            'Batch_ID': n4
        }
    )
studat = pd.DataFrame(studat)
sname = studat['Student_Name'].to_list()
rollno = studat['Roll_Number'].to_list()
batid = studat['Batch_ID'].to_list()
studat.to_csv("Student.csv")

#Course File
file02 = open("Course.csv", "r+")

cd = csv.writer(file02)
coudat = []
for i in range(0,3):
  o1 = input('Course ID : ')
  o2 = input('Course Name : ')
  o3 = {}
  for j in range(0,9):
    ox = int(input(f'Marks Obtained by {sname[i]}: '))
    o3[sname[j]] = ox 
  coudat.append(
    {
   	'Course_ID': o1,
      'Course_Name': o2,
      'Marks Obtained by Students': o3
    }
  )
coudat = pd.DataFrame(coudat)
marstu = coudat['Marks Obtained by Students'].to_list()
marstu = pd.DataFrame(marstu)
counam = coudat['Course_Name'].to_list()
coudat.to_csv('Course Details.csv')

#MarksDistribution
gradlist = []
mardist = []
for j in marstu.index:
  t = marstu[0][j]
  json_acceptable_string = t.replace("'", "\"")
  d = json.loads(json_acceptable_string)
  mardist.extend(d.values())

#Analysis of Grades received by students
sgrade = []
Grading_System = ['A', 'B', 'C', 'D', 'E', 'F', 'G']
for j in range(0,27):
  if (mardist[j] >= 90) and (mardist[j] < 100):
    GRADE = 'A'
    gradlist.append(GRADE)
  if (mardist[j] >= 80) and (mardist[j] < 90):
    GRADE = 'B'
    gradlist.append(GRADE)
  if (mardist[j] >= 70) and (mardist[j] < 80):
     GRADE = 'C'
     gradlist.append(GRADE)
  if (mardist[j] >= 60) and (mardist[j] < 70):
    GRADE = 'D'
    gradlist.append(GRADE)
  if (mardist[j] >= 50) and (mardist[j] < 60):
    GRADE = 'E'
    gradlist.append(GRADE)
  if (mardist[j] >= 40) and (mardist[j] < 50):
    GRADE = 'F'
    gradlist.append(GRADE)
  if (mardist[j] < 40):
    GRADE = 'G'
    gradlist.append(GRADE)

grd_count = []
for i in range(0,7):
  gr = gradlist.count(Grading_System[i])
  grd_count.append(gr)
plt.bar(Grading_System, grd_count)
plt.title('Frequency of students in a particular grade')
plt.xlabel('Grades')
plt.ylabel('Frequency')
plt.show()

#Report Card of a particular student
SNamein = input("Enter the Student's Name you want to fetch marks of:-")
result = sname.count(SNamein)
smar = []
if result > 0:
  for j in marstu.index:
    t = marstu[0][j]
    json_acceptable_string = t.replace("'", "\"")
    d = json.loads(json_acceptable_string)
    smar.append(d[SNamein])
    if (smar[j] >= 40):
      Status = 'Pass'
      if (smar[j] >= 90) and (smar[j] < 100):
        GRADE = 'A'
        sgrade.append(GRADE)
      if (smar[j] >= 80) and (smar[j] < 90):
        GRADE = 'B'
        sgrade.append(GRADE)
      if (smar[j] >= 70) and (smar[j] < 80):
        GRADE = 'C'
        sgrade.append(GRADE)
      if (smar[j] >= 60) and (smar[j] < 70):
        GRADE = 'D'
        sgrade.append(GRADE)
      if (smar[j] >= 50) and (smar[j] < 60):
        GRADE = 'E'
        sgrade.append(GRADE)
      if (smar[j] >= 40) and (smar[j] < 50):
        GRADE = 'F'
        sgrade.append(GRADE)
    else:
      Status = 'Fail'
      GRADE = 'G'
      sgrade.append(GRADE)
else:
  print('INVALID')

totalmar = []
n = 0
for i in range(0,9):
  for j in range(0,3):
    S_C_Marks = mardist[3*j + i]
    n = n + S_C_Marks
  n = n/3
  totalmar.append(n)

x = 0
B_percent = []
for i in range(0,3):
  for j in range(0,3):
    x = x + totalmar[3*i + j]
    B_percent.append(x)

bpercent = []
y = 0
for i in range(0,3):
  for j in range(0,3):
    v = B_percent[3*i + j]
    y = y + v
    y = y/3
  bpercent.append(y)
print(bpercent)


sturep = open('Report Card.txt', 'x+')
if result > 0:
  sturep.write(f"Name: {SNamein}\nPercentage = {totalmar[result]}%\n")
else:
  print("ERROR!")
for i in range(0,3):
  sturep.write(f'{counam[i]}: {smar[i]}\nGrade: {sgrade[i]}\nStatus = {Status}\n')


#Batch File
file03 = open("Batch Details.csv", "x+")

dept_names = ['CSE', 'ECE', 'IT']
BD = csv.writer(file03)
batdat = []
for i in range(0,3):
    q1 = input('Batch_ID: ')
    q4 = counam
    q2 = input('Batch Year : ')
    q3 = []
    for j in range(0,3):
      qx = input('Department Name: ')
      q3.append(qx)
      q5 = sname[3*j + i] 
      BD.append(
        {
            'Batch_ID': q1,
            'Batch_Name': q2,
            'Department_Name': q3,
            'List of Courses': q4,
            'List of Students': q5
        }
    )
batdat = pd.DataFrame(batdat)
dept_names = ['CSE', 'ECE', 'IT']
batname = batdat['Batch_Name'].to_list()
B_yr = batdat['Batch_Year'].to_list()
plt.plot(B_yr, B_percent)
plt.title('Line Plot showing the performance of students batchwise')
plt.xlabel('Batch Year')
plt.ylabel('Average Percentage of Students - Batchwise')
plt.show()

colors = ['red', 'blue', 'green']
for i in range(0,3):  
  for j in range(0,3):
    z = []
    for k in range(0,3):
      z.append(mardist[3*k + j + 9*i])
    plt.scatter(B_yr, z, color = colors[j])  
plt.title('Scatter Plot showing the performance of students in the Annual Exam')
plt.xlabel('Batch')
plt.ylabel('Marks')
plt.show()

batdat.to_csv('Batch Details.csv')

#Department File
file04 = open("Department.csv", "x+")
dd = csv.writer(file04)
deptdat = []
for i in range(0,3):
    p1 = input('Department ID : ')
    p2 = input('Department Name : ')
    p3 = []
    deptdat.append(
        {
            'Dept_ID': p1,
            'Dept_Name': p2,
            'List of Batches': p3
        }
    )
deptdat = pd.DataFrame(deptdat)
deptdat.to_csv('Department.csv')

file05 = open("Annual Examination.csv", "r+")
ed = csv.writer(file05)
examdat = []
for i in range(0,3):
  for j in range(0,3):
      r2 = sname[3*i + j]
      for k in range(0,3):
          r1 = batname[j]
          r3 = counam[k]
          r4 = mardist[9*k + 3*i + j]
          examdat.append(
              {
                  'Batch_Name': r1,
                  'Student Name': r2,
                  'Subject': r3,
                  'Marks': r4
              }
          )
examdat = pd.DataFrame(examdat)
examdat.to_csv("Annual Examination.csv")
```
